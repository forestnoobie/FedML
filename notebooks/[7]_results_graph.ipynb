# Fedavg vs FedCon

```python
# Alpha 기준으로 그래프 그리고
# 한번은 epoch, 한번은 client ratio 기준으로 그리자
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.size': 20})
```

```python
# data 설정 Client number 8 / Total 20 clients

epoch_x = ["5", "10", "20"]
epoch_y = {}
epoch_y['Fedavg'] = {}
epoch_y['FedCon'] = {}

epoch_y['Fedavg'][0.01] = [22.5, 22.43, 22.76]
epoch_y['Fedavg'][0.05] = [42.74, 40.64, 39.53]
epoch_y['Fedavg'][100] = [70.74, 69.68, 68]


epoch_y['FedCon'][0.01] = [22.39, 24.18, 23.18]
epoch_y['FedCon'][0.05] = [47.36, 44.69, 41.16]
epoch_y['FedCon'][100] = [69.61, 66.99, 66.66]
```

```python
alpha = 0.01
epoch_x = ["5", "10", "20"]

  
# plot lines
plt.plot(epoch_x, epoch_y['Fedavg'][alpha], label = "Fedavg", marker="o")
plt.plot(epoch_x, epoch_y['FedCon'][alpha], label = "FedCon", marker="+")

plt.figsize=(6, 3)
plt.ylim(15, 25)
plt.ylabel("Top-1 test accuracy")
plt.xlabel("# of local epochs")
plt.legend()
plt.tight_layout()
plt.savefig("./7_results/avg_con_epochs_{}.pdf".format(alpha), bbox="tight")

plt.show()
```

```python
alpha = 0.05
  
# plot lines
plt.plot(epoch_x, epoch_y['Fedavg'][alpha], label = "Fedavg", marker="o")
plt.plot(epoch_x, epoch_y['FedCon'][alpha], label = "FedCon", marker="+")

plt.figsize=(6, 3)
plt.ylim(38, 48)
plt.ylabel("Top-1 test accuracy")
plt.xlabel("# of local epochs")
plt.legend()
plt.tight_layout()
plt.savefig("./7_results/avg_con_epochs_{}.pdf".format(alpha), bbox="tight")
plt.show()
```

```python
alpha = 100

  
# plot lines
plt.plot(epoch_x, epoch_y['Fedavg'][alpha], label = "Fedavg", marker="o")
plt.plot(epoch_x, epoch_y['FedCon'][alpha], label = "FedCon", marker="+")

plt.figsize=(6, 3)
plt.ylim(61, 71)
plt.ylabel("Top-1 test accuracy")
plt.xlabel("# of local epochs")
plt.legend()
plt.tight_layout()
plt.savefig("./7_results/avg_con_epochs_{}.pdf".format(alpha), bbox="tight")
plt.show()
```
